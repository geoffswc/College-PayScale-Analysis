```python
import pandas as pd
import requests
```

```python
# didn't automate, just got these by clicking through each one on
#https://www.payscale.com/college-salary-report/best-schools-by-majors/
# note - no data for business or education, not sure why?

majors = {'humanities':22, 'computer-science':25, 'social-sciences':34, 'art':17, 'engineering':17, 
         'math':5, 'communication':18, 'physical-and-life-sciences':21,
         'health-science':27}
```

```python
all_pages = []
for m in majors.keys():
    for i in range(1,majors[m]):
        url = 'https://www.payscale.com/college-salary-report/best-schools-by-majors/' + m + '/page/' + str(i)
        print(url)
        df = pd.read_html(url)
        df = df[0].iloc[0:,1:]
        df.columns = ['school','early','mid']

        #df['school'] = df['school'].str.split(':')[1]
        df.school = df.school.str.split(':').str[1]
        df.early = df.early.str.split(':').str[1]
        df.mid = df.mid.str.split(':').str[1]
        df['major'] = m
        all_pages.append(df)
```

```python
df = pd.concat(all_pages)
```

```python
df = df.reset_index()
```

```python
df.early = df.early.str.replace('$','')
df.early = df.early.str.replace(',','')
df.mid = df.mid.str.replace('$','')
df.mid = df.mid.str.replace(',','')
df.early = pd.to_numeric(df.early, errors='coerce')
df.mid = pd.to_numeric(df.mid, errors='coerce')
```
