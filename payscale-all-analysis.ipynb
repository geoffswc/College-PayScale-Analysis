```python
import pandas as pd
```

```python
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
```

```python
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('payscale-all-majors-colleges.csv')
```

```python
df = df.reset_index()
```

```python
pysqldf("SELECT * FROM df ORDER BY late DESC LIMIT 100")
```

```python
pysqldf("SELECT * FROM df ORDER BY early DESC LIMIT 100")
```

```python
df.groupby("major").describe()['late']
```

```python
df.groupby("major").describe()['early']
```

```python
# the average ranking
```

```python
pysqldf("""
WITH dsrank AS 
(
SELECT
    school, major, late, ROW_NUMBER() OVER (ORDER BY late DESC) AS ranking
FROM
    df
)

SELECT 
    major, AVG(ranking)
FROM
    dsrank
GROUP BY 
    major
""")
```

```python
pysqldf("""
WITH dsrank AS 
(
SELECT
    school, major, late, ROW_NUMBER() OVER (ORDER BY late DESC) AS ranking
FROM
    df
LIMIT 20
)

SELECT 
    major, AVG(ranking)
FROM
    dsrank
GROUP BY 
    major
""")
```

```python
# list the top 10 of each major, in order of late career salary
pysqldf("""
SELECT * FROM
(
SELECT
    school, major, late, ROW_NUMBER() OVER (PARTITION BY major ORDER BY late DESC) AS ranking
FROM
    df
)
WHERE 
    ranking <= 10
ORDER BY 
    late DESC
""")
```
