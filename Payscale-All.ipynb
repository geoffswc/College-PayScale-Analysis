```python
import pandas as pd
import requests
```

```python
majors = {'humanities':23, 'computer-science':25, 'social-sciences':35, 'art':17, 'engineering':17, 
         'math':6, 'business':49, 'education':14, 'communication':19, 'physical-and-life-sciences':22,
         'health-science':28}
```

```python
all_pages = []
for m in majors.keys():
    for i in range(1,majors[m]):
        url = 'https://www.payscale.com/college-salary-report/best-schools-by-majors/' + m + '/page/' + str(i)
        print(url)
        df = pd.read_html(url)
        df = df[0].iloc[0:,1:]
        df.columns = ['school','early','late']

        #df['school'] = df['school'].str.split(':')[1]
        df.school = df.school.str.split(':').str[1]
        df.early = df.early.str.split(':').str[1]
        df.late = df.late.str.split(':').str[1]
        df['major'] = m
        all_pages.append(df)
```

```python
df = pd.concat(all_pages)
```

```python
df = df.reset_index()
```

```python
df.early = df.early.str.replace('$','')
df.early = df.early.str.replace(',','')
df.late = df.late.str.replace('$','')
df.late = df.late.str.replace(',','')
df.early = pd.to_numeric(df.early, errors='coerce')
df.late = pd.to_numeric(df.late, errors='coerce')
```
