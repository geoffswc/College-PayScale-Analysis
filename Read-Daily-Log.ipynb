```python
df = pd.read_html(r.text)
```

```python
month_year = ['may-2020', 'april-2020', 'march-2020', 'february-2020', 'january-2020']
```

```python
daily_logs = []
for m in month_year:
    r = requests.get('https://police.ucsf.edu/crime-prevention-statistics/daily-crime-logs/' + m)
    daily_logs.append(pd.read_html(r.text)[0])
```

```python
len(daily_logs)
```

```python
# vertically stack all the data frames for each month
df = pd.concat(daily_logs)
```

```python
df
```

```python
df['Reported'] = pd.to_datetime(df['Reported'])
```

```python
df['month'] = df['Reported'].dt.month
```

```python
import matplotlib.pyplot as plt
```

```python
df.groupby(df.month).count()['Reported'].plot(kind="bar")
```
